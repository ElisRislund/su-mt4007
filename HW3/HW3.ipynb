As we can see, the proportion of people who have at some point been married is signficantly much higher in the stroke group compared to the non-stroke group. Why could this be? Does getting married make you more likely to suffer from a stroke? One way to attempt to explain why could be with the following two frequency plots: 

```r
ggplot(df_stroke,aes(x=age))+
    geom_freqpoly(aes(color = stroke))+
labs(title = "Frequency polygon 2: Age and stroke",
    x = "Age",
    y = "Frequency")

ggplot(df_stroke,aes(x=age))+
    geom_freqpoly(aes(color = ever_married))+
labs(title = "Frequency polygon 3: Age and married",
    x = "Age",
    y = "Frequency")
```

As we can see from the Frequency plot 2 and 3, there is a trend that people with stroke and people who have been married at some point are older. This is very natural. Both having had a stroke and being married at some point are cumulative; if you have never had a stroke or been married you can either stay that way or get a stroke or get married - but once you have gotten a stroke or been married there is no going back to no stroke or never been married. For this reason it is only natural that people with stroke and ever been married are older in general, which in turn automatically results in the kind of positive correlation shown in Barplot 2, which does not necessarily indicate any causation (for instance getting married increasing your risk of stroke). This effect is further amplified since we found that our data included more old people than young people. So whilst the difference in proportion found in Barplot 2 was very large it might totally insignificiant and just a statistical inevitability from how our variables were defined. On the other hand, whilst the difference in proportion found in Barplot 1 was much smaller it might actually say something significant since gender is not very clearly correlated with age (cf Frequency polygon 1). ## Task 4: Data preperation

We have several different unusual values in our data. We have some values that are zero, some that are NA, and some that are empty strings. 

```r
df_prep <- read.csv("https://raw.githubusercontent.com/su-mt4007/data/refs/heads/main/cell_phones_total.csv")
df_prep
```

Mobile phones were invented in the 1970s. For this reason there is no practical reason for us to have values for 
the year 1960, i.e. include the year 1960 as a variable. Moreover how should one even interpret a zero vs an NA in our data? Since mobile phones were not invented yet we can safely say that the all values in 1960 column should be zero. However it does not seem reasonable that we should treat all NAs as zeroes. At some point during the 1970s some countries began to have a non-zero number of mobile phones in them, but mobile phones were not yet widely spread. This introduces some uncertainty: how should we decide which NAs and zeroes should stand for literally zero mobile phones and which zeroes and NAs should stand for missing values? This is an impossible task, and we must pick some reasonable method to circumvent this. One such method is to only consider data after when mobile phones got sufficiently widespread. Mobile phones got globally popular in the 1990s and 2000s, so reasonable we should remove all time variables before this. To decide upon the initial year I looked at the proportion of countries that had zeros or NAs as their values for different years. If the proportion was sufficiently low and if the year-to-year change in proportion (comparing with adjacent years) was low I thought that it would be a sound starting point. I ended up settling for the year 2000. This year the proportion of countries with NAs and zeroes was less than 5 percent and the change from previous years was quite small. 

```r
sum(df_prep$X2000 == 0 | is.na(df_prep$X2000))/214
df_post_2000 <- select(df_prep,-2:-37)
sum(is.na(df_post_2000))
```

In particular we note that out of all variables (years 2000 to 2019) there are no NAs left in our data. Since this was a point when mobile phones were quite wide spread and a topic of general interest (so measuring number of mobile phones in a country without mobile phones was still a reasonable thing to do) I interpreted all zeros in the data as literal zeroes (not stand-ins for missing values). However we still have some missing values in the form of empty springs. I found it reasonable to fill these in using last observation carried forward. A problem was that some countries had a lot of missing values. To deal with these countries I simply dropped them if they had too many. I chose too many to mean more than 10 missing values. 10 missing values in the period 2000 to 2019 would be 50 percent missing, which is very high and would make any interpolation bad.

```r
df_post_2000_2 <- filter(df_post_2000,rowSums(df_post_2000 == "") < 10)
```

Then I carried on by doing a locf and locb row-wise. Since all rows have atleast one observation, this is going to fill in the dataframe.

```r
# Changing empty strings to NA's for locf and locb
df_mutated <- mutate(df_post_2000_2,across(everything(), ~ na_if(., "")))

# Removing id row for countries
first_column <- df_mutated[, 1, drop = FALSE]


df_final <- as.data.frame(
  cbind(
    first_column,  # Add the id row back for countries to the locf + locb data frame
    t(apply(df_mutated[, -1], 1, function(row) zoo::na.locf(zoo::na.locf(row, na.rm = FALSE), fromLast = TRUE)))
  )
)

convert_to_numeric <- function(x) {
  if (str_detect(x, "k")) {
    return(as.numeric(str_remove(x, "k")) * 1e3)
  } else if (str_detect(x, "M")) {
    return(as.numeric(str_remove(x, "M")) * 1e6)
  } else if (str_detect(x, "B")) {
    return(as.numeric(str_remove(x, "B")) * 1e9)
  } else {
    # If no suffix, just return the numeric value
    return(as.numeric(x))
  }
}

# Apply our converting function defined above to each column except
# the first one which is the id for country:
df_final2 <- df_final |>
  mutate(across(
    .cols = -1,  
    .fns = ~sapply(., convert_to_numeric)
  ))
print("Tal istället för strängar")
df_final2
```

At this point we have dataframe with no missing values (all values were filled in using locf then locb. Now we just convert all the data stored as strings to numerical values:

```r
convert_to_numeric <- function(x) {
  if (str_detect(x, "k")) {
    return(as.numeric(str_remove(x, "k")) * 1e3)
  } else if (str_detect(x, "M")) {
    return(as.numeric(str_remove(x, "M")) * 1e6)
  } else if (str_detect(x, "B")) {
    return(as.numeric(str_remove(x, "B")) * 1e9)
  } else {
    return(as.numeric(x))
  }
}

# Apply our converting function defined above to each column except
# the first one which is the id for country:
df_final2 <- df_final |>
  mutate(across(
    .cols = -1,  
    .fns = ~sapply(., convert_to_numeric)
  ))
df_final2
```
